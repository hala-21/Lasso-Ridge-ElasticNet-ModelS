```python
#importing data frame and setting the columns names
import pandas as pd

url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data'
column_names = ['MPG', 'Cylinders', 'Displacement', 'Horsepower', 'Weight',
                'Acceleration', 'Model Year', 'Origin']
car_df = pd.read_csv(url, names=column_names,
                          na_values='?', comment='\t',
                          sep=' ', skipinitialspace=True)
car_df
```

```python
#droping all nulls
car_df=car_df.dropna()
car_df

#normalizing data
from sklearn.preprocessing import MinMaxScaler
ncar_df=MinMaxScaler().fit_transform(car_df)
ncar_df
```

```python
#put values in descrite form
Ncar_df=pd.DataFrame(ncar_df,columns=car_df.columns)
Ncar_df
```

```python
#setting iutput values or targeted values to i
i = Ncar_df.drop(labels=["MPG"],axis=1)
i
```

```python
#setting onput values to i
o=Ncar_df[['MPG']]
o
```

```python
# first let's see the code without  the polynomial features
```

```python
#applay the train_test_split to our input and output
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(i, o, test_size=0.3, random_state=42)
X_test
```

```python
#applaying linear regression to a model
from sklearn.linear_model import LinearRegression
#lin_Reg_model=LinearRegression()
#lin_Reg_model.fit(X_train,y_train)
lin_Reg_model=LinearRegression().fit(X_train,y_train)
print('LinearRegression Model Training accuracy is : ' + str( lin_Reg_model.score(X_train,y_train)))

print('LinearRegression Model Testing accuracy is :' +str( lin_Reg_model.score(X_test,y_test)))
```

```python
#applaying linear regression to a model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
lin_Reg_model=LinearRegression().fit(X_train,y_train)
y_pred=lin_Reg_model.predict(X_train)
mse=mean_squared_error(y_train,y_pred)
print('Trainng error : '+ str(mse))
y_preds=lin_Reg_model.predict(X_test)
mses=mean_squared_error(y_test,y_preds)
print('Testing error : '+str(mses))
```

```python
# now let's start with the lasso model
from sklearn.linear_model import Lasso
lasso_model=Lasso(alpha=0.1)
lasso_model.fit(X_train,y_train)

print('Lasso Model Training accuracy is : ' + str( lasso_model.score(X_train,y_train)))

print('Lasso Model Testing accuracy is : ' + str( lasso_model.score(X_test,y_test)))
```

```python
# now let's start with the ridge model
from sklearn.linear_model import Ridge
ridge_model=Ridge(alpha=0.1)
ridge_model.fit(X_train,y_train)

print('Ridge Model Training accuracy is : ' + str( ridge_model.score(X_train,y_train)))

print('Ridge Model Testing accuracy is : ' + str( ridge_model.score(X_test,y_test)))
```

```python
# now let's start with the elisticnet model
from sklearn.linear_model import  ElasticNet
ElasticNet_model= ElasticNet(alpha=0.1)
ElasticNet_model.fit(X_train,y_train)

print('ElasticNet Model Training accuracy is : ' + str(  ElasticNet_model.score(X_train,y_train)))

print('ElasticNet Model Testing accuracy is : ' + str(  ElasticNet_model.score(X_test,y_test)))
```
